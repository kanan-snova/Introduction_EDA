### 3. データの数値化して概要を知る

```python
# 並べ替えをしてみよう
# データフレーム.sort_values(by=['並べ替え基準の変数名'], ascending=並べ順)
# ラーメン店舗数が一番多いのはどの県？（ラーメン店舗の多い順に並び替えてみよう）
df.sort_values(by=['ramen_shop'], ascending=False)  #ascending=Trueにすると昇順
```

```python
#　要約統計量を算出しよう
# count 件数
# mean 平均
# std 標準偏差
# min 最小値
# 25% 25%点
# 50% 50%点（中央値）
# 75% 75%点
# max 最大値
df.describe()
```

```python
# 新しい列を計算して作成する
# 女性比率を求める（各県の女性の人口÷各県の人口）
# 計算結果を'female_%'という名前の変数を新しく作成する
df['female_%'] = df['female_k']/df['population_k']
```

```python
# 先頭5行を表示して、一番最後（右端）に新しい変数が追加されていることを確認する
df.head()
```

```python
# 度数集計をする
# 地方ごとに幾つの県があるかカウントする
# value_counts()関数はカウント数降順で結果が表示される
df['地方'].value_counts()
```

```python
# グルーピングする
grouped = df.groupby('地方')
#　グループ化した「地方」別に、ラーメン店の数の要約統計量を算出してみる
grouped['ramen_shop'].describe()
```

```python
# グループ化した「地方」別に、ソースの要約統計量を算出してみる
grouped['source_ml'].describe()
```

```python
# １行でグループ化して要約統計量算出を書く方法もある（ご参考）
df.groupby('地方')['ramen_shop'].describe()
```

```python
# 1つの指標だけ求めることもできる
# 今回は地方ごとのラーメン店の平均を表示
df.groupby('地方')['ramen_shop'].mean()
```

### 4. matplotlibを使って可視化（グラフ化）

```python
# グラフを書く時にメインで使うのは matplotlibというライブラリ
# seabornは綺麗なグラフが描けたり、細かい設定でグラフ描けたりもするライブラリ
import matplotlib.pyplot as plt
import seaborn as sns

# 日本語文字化け対応用ライブラリをインポートする
import japanize_matplotlib

# おまじない↓（結果を実行した真下にちゃんと表示させるためにやります）
%matplotlib inline
```

```python
# 地方別の面積の棒グラフ
# plt.bar(x軸の変数, y軸の変数)
# x軸はカテゴリ変数(文字列)、y軸は数値変数が層別棒グラフの基本
plt.bar(df['地方'],df['area_km2'])
plt.show()
```

```python
# 地方別の人口の棒グラフ
# 面積の棒グラフと比較して、人口の棒グラフはどうなっているか考えてみよう
plt.bar(df['地方'],df['population_k'] ,color='pink') # color=で色変更できる
plt.show()
```

```python
# ラーメン店のヒストグラムを出してみる
# ヒストグラムは数値変数のグラフ化方法
plt.hist(df['ramen_shop'])
plt.show()
# ビンの数（分割数）を指定するとグラフの印象が変わる
# bins=で指定する
plt.hist(df['ramen_shop'], bins=30)
plt.show()
```

```python
# タイトルやx軸、y軸の凡例をつけたり、大きさ変えたりできます
plt.figure(figsize=(12,5))
plt.hist(df['ramen_shop'], bins=20, color='green', alpha=0.5)
plt.title('ラーメン屋店舗数のヒストグラム')
plt.ylabel('都道府県数')
plt.xlabel('ラーメン店舗数')
plt.show()
```

```python
#平均値を確認してみる
df['ramen_shop'].mean()
```

```python
#中央値を確認してみる
df['ramen_shop'].median()
```

```python
# ヒストグラムだと平均値や中央値がどのあたりにあるかわかりづらい
# なので箱ひげ図を書いてみる
plt.boxplot(df['ramen_shop'],showmeans=True)
plt.show()
```

```python
# 地方別に箱ひげ図を描いてみよう
# 層別での箱ひげ図はseabornライブラリを使う
# 箱ひげ図を横に並べると分布がわかりやすい
sns.boxplot(x='地方', y='ramen_shop',data=df)
plt.show()
```

```python
# 散布図は数値変数と数値変数の２変数の関係性を表すグラフ
# 人口とラーメン店の関係性を散布図でみてみる
# alpha=0.5は色の透過の設定なので、見栄えの問題
plt.scatter(df['population_k'], df['ramen_shop'], alpha=0.5)
plt.show()
```
